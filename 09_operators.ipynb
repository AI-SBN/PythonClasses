```python
a=10
b=20
print("Before Swapping a=",a)
print("Before Swapping b=",b)
a=a-b  # a = 10 - 20 = -10
b=a+b  # b = -10+20 = 10
a=b-a # a = 10-(-10) = 20
print("After Swapping a=",a)
print("After Swapping b=",b)
```
